```python
def compute_steps(p_data):
    n_steps = 0
    prev_loc = (p_data[0]['position'][1], p_data[0]['position'][0])
    for i in range(1, len(p_data)):
        new_loc = (p_data[i]['position'][1], p_data[i]['position'][0])
        if new_loc != prev_loc:
            n_steps += 1
            prev_loc = new_loc
    return n_steps
    
    
    
```
